```python
import mysql.connector
```

```python
connection = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    password = 'archit',
    database = 'porches'
)
```

```python
cursor = connection.cursor(dictionary = True)
```

```python
cursor.execute("SHOW TABLES;")
```

```python
tables = cursor.fetchall()
```

```python
tables
```

```python
tables[0]['Tables_in_porches']
```

```python
schema = {}
```

```python
for table in tables:
    table_name = list(table.values())[0]
    desc_cursor = connection.cursor(dictionary = True)
    desc_cursor.execute(f"DESCRIBE {table_name};")
    schema[table_name] = desc_cursor.fetchall()
    desc_cursor.close()
```

```python
schema['students']
```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(
    model = "openai/gpt-oss-120b",
)
```

```python
from langchain_core.prompts import ChatPromptTemplate
system_prompt = "You are a sql assistant which is connected to execute_query agent so you must return the schema directly"
user_question = "what is the schema of db students"
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    ("human", "{user_input}")
])
```

```python
chain = prompt | llm
response = chain.invoke(f"{"user_input": user_question, "schema": {schema}}", tools = None)
```
